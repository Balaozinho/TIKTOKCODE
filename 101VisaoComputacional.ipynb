```python
import openai
from openai import OpenAI
from dotenv import load_dotenv, find_dotenv
import base64

_ = load_dotenv(find_dotenv())
client = OpenAI()
```

```python
#criar uma funcao de encode
def encode_image(caminho_imagem):
    with open(caminho_imagem, 'rb') as img:
        return base64.b64encode(img.read()).decode('utf-8')

caminho = 'ExemploImagem.jpeg'
base_64_img = encode_image(caminho)

texto = 'Me ajude a entender o que está escrito na imagem nas anotações do caderno de estudo em Python'

resposta = client.chat.completions.create(
    model='gpt-4o',
    messages=[
        {
            "role": "user",
            "content": [
                {"type": "text", "text": texto},
                {
                    "type": "image_url",
                    "image_url": {
                        "url": f"data:image/jpeg;base64,{base_64_img}"
                    }
                }
            ]
        }
    ],
)

print(resposta.choices[0].message.content)
```
