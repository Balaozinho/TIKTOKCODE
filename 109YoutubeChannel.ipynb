```python
import pandas as pd
import numpy as np

df = pd.read_csv("top-300-youtube-channels.csv")
df.head(5)
```

```python
#RELAÇÃO #INSCRITOS / #VIDEOS POSTADOS
df['subs_per_video'] = np.where(
    df['Video_Count'] > 0,
    df['Subscriber_Count'] / df['Video_Count'], 0
)

df[['Channel_Name', 'Genre', 'Video_Count', 'subs_per_video']] \
    .sort_values(by='subs_per_video', ascending=False) \
    .head(10).round(2).style.format({'subs_per_video' : '{:,.2f}'})
```
