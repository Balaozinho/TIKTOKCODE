```python
import pandas as pd
import numpy as np

df = pd.read_csv('top-300-youtube-channels.csv')
df.head(5)
```

```python
df['views_per_video'] = np.where(
    df['Video_Count'] > 0,
    df['Video_Views'] / df['Video_Count'],
    0
)

high_impact_channels = (
    df[['Channel_Name', 'Genre', 'views_per_video']]
    .sort_values(by='views_per_video', ascending=False)
    .head(10).round(2).style.format({'views_per_video':'{:,.2f}'})
)

high_impact_channels
```
